### Type of Variable Names
1. <b>Camel Case </b>: Except for the first word, each word starts with a capital letter in Camel Case.
For example, myAge.
2. Pascal Case: Each word of the variable name starts with a capital letter in Pascal Case.
For example,  MyAge.
3. Snake Case: Each word of the variable name is separated by an underscore in a Snake case.
For example, my_age.

#### Global dan Local Variable

Local and global variables are defined based on the position where it is declared. This is very important to understand these two terms and be able to differentiate if a variable is local or global.

Local Variable:
If a variable is declared inside the function, it is called a local variable. We cannot call local variables outside the function because they are defined and declared inside it.

#### Global variable dan local variable
<img src="Global-local var.png" width="500" height="240">

```python
# Membuat fungsi
x = 20
def func():
    y = 25
    global x
    x = y+30    
func()
print('Fungsi variabel lokal dan global:', x)
```

### Metode string

String adalah jenis yang paling populer di bahasa pemrograman. Kita bisa membuatnya hanya dengan melampirkan karakter dalam tanda kutip. Python memperlakukan tanda kutip tunggal sama dengan tanda kutip ganda. Metode string adalah kumpulan fungsi bawaan yang beroperasi pad python.

| INPUT | METODE | OUTPUT |
|----| ----| ---|
|"hello world"| <mark> capitalize() </mark> | Hello world|
|"hello world"| <mark> .isalpha() </mark> | True|
|"12345"| <mark> .isnumeric() </mark> | True|
|"hello world"| <mark> .isupper() </mark> | False|
|"Hi Alam"| <mark> .split() </mark> | "Hi", "Alam"|
|"hello world"| <mark> .title() </mark> | Hello World|
|" hello "| <mark> .strip() </mark> | Hello|
|" x y z "| <mark> .replace('x', 'A') </mark> | "A y z"|

```python
var = 'hello world'
var1 = 'helloworld'
print(var.capitalize())
print(var.isalpha())
print(var.isupper())
print(var.split())
```

```python
#Covert list to tuple

lst = [[1, 2], [3, 4], [5, 6]] 
tuples = [tuple(x) for x in lst] 
print(tuples) 
# [(1, 2), (3, 4), (5, 6)]
```

```python
# DICTIONARY
teman = [['Aco', 25, 'pirang'], 
['Baso', 33, 'black'], 
['Cikal', 18, 'abu-abu']] 
teman_dict = dict((x[0], x[1:]) for x in teman) 
print(teman_dict) 
```
